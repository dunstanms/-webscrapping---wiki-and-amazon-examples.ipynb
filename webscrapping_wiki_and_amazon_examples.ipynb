```python
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Kenya"
```

```python
#Package the request, send the request and catch the response: r
r = requests.get(url)
```

```python
# Extract the response as html: text
text = r.text
```

```python
text
```

```python
from bs4 import BeautifulSoup
```

```python
#Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(text,"lxml")
```

```python
soup #prettified soup
```

```python
type(soup)
```

```python
print(soup.title.text)
```

```python
print(soup.get_text())
```

```python
# Find all 'a' tags (which define hyperlinks)
a_tags = soup.find_all('a')
```

```python
a_tags
```

```python
soup.find_all(class_="thumb tright")
```

```python
# Print the URLs

for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
url = requests.get("https://www.amazon.in/Bose-SoundTouch-Wireless-Music-System/dp/B014SKZNEA/ref=sr_1_1?s=electronics&ie=UTF8&qid=1526745651&sr=1-1&keywords=bose+soundtouch+20")
```

```python
soup = BeautifulSoup(url.content, 'html.parser')
soup
```

```python
review1 = soup.findAll(class_="a-spinner-wrapper cr-reviews-loading aok-hidden")
review1
```

```python
for article in review1:
    print (article.text.strip(),"\n")
```

```python
reviews=[]
for article in review1:
    reviews.append(article.text.strip())
```

```python
reviews
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(reviews)
```

```python
df
```

```python
url = requests.get("https://www.amazon.in/Bose-SoundTouch-Wireless-Music-System/product-reviews/B014SKZNEA/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews")
```

```python
review1 = soup.findAll(class_="a-icon a-icon-star a-star-5 review-rating")
```

```python
for article in review1:
    print(article.text,"\n")
```
